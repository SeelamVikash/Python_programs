```python
cards = ["ACE",2,3,4,5,6,7,8,9,10,"jack","queen","king"]
suits = ["Clubs","Spades","Diamonds","Hearts"]
```

```python
#deckclass this class is used to create deck of cards and shuffle them
class Deck:
    def __init__(self,cards,suits):
        self.cards=cards
        self.suits=suits
        self.deck = []
        for i in self.suits:
            for j in self.cards:
                self.deck.append(f"{j} of {i}")            
    def __str__(self):
        return self.deck
    
    def Shuffle(self):
        '''
        used to shuffle the deck of cards
        '''
        random.shuffle(self.deck)       
        
```

```python
#class for Bet monitoring

class Bet:
    def __init__(self,bet,balance):
        self.bet =bet
        self.balance = balance
    def win(self):
        '''
        use it to calculate the win amount
        '''
        self.balance+=self.bet
        return self.balance
    def loss(self):
        '''
        use this calculate loss
        '''
        self.balance-=self.bet
        return self.balance
```
